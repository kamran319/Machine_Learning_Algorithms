```python
label_encoder=preprocessing.LabelEncoder()
```

# Model_1

```python
tree_model=tree.DecisionTreeClassifier()
```

```python
tree_model.fit(X=pd.DataFrame(encoded_sex),y=train['Survived'])
```

```python
with open("DecisionTree.dot","w")as f:
    f=tree.export_graphviz(tree_model,feature_names=['Sex'],out_file=f);
```

# Model_2

```python
tree_model.fit(pd.DataFrame([encoded_sex,train['Pclass']]).T,y=train['Survived'])
```

```python
with open("DecisionTree1.dot","w") as f:
    f=tree.export_graphviz(tree_model,feature_names=["Sex","Pclass"],out_file=f);
```

# Model_3

```python
predictors=pd.DataFrame([encoded_sex, train["Pclass"], train["Age"],train["Fare"]]).T
```

```python
# 8 depth has given here as we have 4 independent variable and 2 catergories of depenedent varibale 
#i.e syrvived or not thats why 4*2=8 has given
tree_model=tree.DecisionTreeClassifier(max_depth=8)
```

```python
tree_model.fit(X=predictors,y=train['Survived'])
```

```python
with open("DecisionTree3.dot","w") as f:
    f=tree.export_graphviz(tree_model,feature_names=["Sex","Pclass","Age","Fare"],out_file=f);
```

# How to check the model accuracy

```python
tree_model.score(X=predictors,y=train['Survived'])
```

# Random Forest

```python
train.columns
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
train['Sex']=label_encoder.fit_transform(train["Sex"])
```

```python
train['Embarked']=label_encoder.fit_transform(train["Embarked"])
```

```python
train.head()
```

```python
train.Embarked
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(n_estimators=10000, max_features=2, oob_score=True)
```

```python
rf_model
```

```python
features=["Sex","Age","Embarked","Pclass","SibSp","Fare"]
```

```python
features
```

```python
rf_model.fit(X=train[features],y=train["Survived"])
```

```python
rf_model.oob_score_
```

# To find important of each independent Varibale 

```python
for features,imp in zip(features,rf_model.feature_importances_):
    print(features,imp)
```
