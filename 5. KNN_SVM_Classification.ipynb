```python
new_age=np.where(train_knn['Age'].isnull(),32,train_knn['Age'])
```

```python
train_knn['Age']=new_age
```

```python
Le=preprocessing.LabelEncoder()
```

```python
train_knn['Sex']=Le.fit_transform(train_knn['Sex'])
```

```python
train_knn['Embarked']=Le.fit_transform(train_knn['Embarked'])
```

```python
train_knn.head()
```

```python
train_knn.drop(['Cabin'],axis=1,inplace=True)
```

```python
train_knn.drop(['Name'],axis=1,inplace=True)
```

```python
train_knn.drop(['Ticket'],axis=1,inplace=True)
```

```python
train_knn.head()
```

```python
from sklearn import neighbors
```

```python
train_knn.drop(['PassengerId'],axis=1,inplace=True)
```

```python
y=train_knn["Pclass"]
X=train_knn.drop(["Pclass"],axis=1)
```

```python
train_knn.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

# KNN

```python
knn=neighbors.KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train)
```

```python
ypred=knn.predict(X_test)
```

```python
accuracy_score(y_test,ypred,normalize=True)
```

```python
confusion_matrix(y_test,ypred)
```

# SVM

```python
from sklearn import svm
```

```python
y=train_knn["Pclass"]
X=train_knn.drop(["Pclass"],axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
clf=svm.SVC(gamma=0.01,C=100)
```

```python
clf.fit(X_train,y_train)
```

```python
ypred=clf.predict(X_test)
```

```python
accuracy_score(y_test,ypred,normalize=True)
```

```python
confusion_matrix(y_test,ypred)
```
